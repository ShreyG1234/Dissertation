Calculation of PMFs

```python
import numpy as np

outcomes = np.array([1, 2, 3, 4, 5, 6])

pmf_values = np.ones_like(outcomes) / len(outcomes)

print("The Outcomes are:", outcomes)
print("The PMF Values are:", pmf_values)
```

Graphing PMFs

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 10
p = 0.3

# Generate values for the binomial PMF
x = np.arange(0, n+1)
pmf_values = binom.pmf(x, n, p)

# Plot the PMF
plt.bar(x, pmf_values)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Probability Mass Function')
plt.show()
```

Simulation of 1000 rolls

```python
num_rolls = 1000
outcomes = np.array([1, 2, 3, 4, 5, 6])
simulated_rolls = np.random.choice(outcomes, size=num_rolls, replace=True)

outcome_counts = np.bincount(simulated_rolls)[1:]

empirical_pmf = outcome_counts / num_rolls

plt.bar(outcomes, empirical_pmf, align='center', alpha=0.7)
plt.xlabel('Outcomes')
plt.ylabel('Empirical Probability')
plt.title('Empirical Probability Mass Function of a Fair Die (from Simulation)')
plt.xticks(outcomes)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
